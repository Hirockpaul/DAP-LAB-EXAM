```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("tips.csv")
```

```python
df.head()
```

```python
df["percentTip"] = df["tip"]/df["total_bill"]*100
print(df.head())
```

```python
px = px.scatter(data_frame = df, x="total_bill", y="percentTip", size="size", color= "day", trendline="ols")
px.show()
```

```python
import plotly.express as px
data2 = df.copy(deep=True)
data2 = data2.groupby('day', as_index=False)['percentTip'].mean()
fig = px.bar(data_frame = data2, x="day", y="percentTip", color= "day")
fig.show()
```

```python
fig = px.pie(df, values='tip', names='day',hole = 0.5)
fig.show()
```

```python
data2 = df.copy(deep=True)
data2 = data2.groupby('sex', as_index=False)['percentTip'].mean()
fig = px.bar(data_frame = data2, x="sex", y="percentTip", color= "sex")
fig.show()
```

```python
data2 = df.copy(deep=True)
data2 = data2.groupby('smoker', as_index=False)['percentTip'].mean()
fig = px.bar(data_frame = data2, x="smoker", y="percentTip", color= "smoker")
fig.show()
```

```python
data2 = df.copy(deep=True)
data2 = data2.groupby('time', as_index=False)['percentTip'].mean()
fig = px.bar(data_frame = data2, x="time", y="percentTip", color= "time")
fig.show()
```

```python
fig = px.pie(df, values='tip', names='time',hole = 0.5)
fig.show()
```

```python
data2 = df.copy(deep=True)
data2 = data2.groupby('size', as_index=False)['percentTip'].mean()
fig = px.bar(data_frame = data2, x="size", y="percentTip", color= "size")
fig.show()
```

```python
#

data2 = df.copy(deep=True)
data2 = data2.groupby('size', as_index=False)['tip'].mean()
fig = px.bar(data_frame = data2, x="size", y="tip", color= "size", labels={"tip": "average tip"})
fig.show()
```

```python
df["sex"] = df["sex"].map({"Female": 0, "Male": 1})
df["smoker"] = df["smoker"].map({"No": 0, "Yes": 1})
df["day"] = df["day"].map({"Thur": 0, "Fri": 1, "Sat": 2, "Sun": 3})
df["time"] = df["time"].map({"Lunch": 0, "Dinner": 1})
df.head()
```
